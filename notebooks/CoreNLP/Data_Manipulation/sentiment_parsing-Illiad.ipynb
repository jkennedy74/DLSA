```python
# Define properties for coreNLP to execute calls
props={'annotators': 'sentiment','pipelineLanguage':'en','outputFormat':'json'}
```

```python
# Define the target filepath
filename = "./Corpi/illiad.txt"
# Read in target file
raw_text = open(filename, encoding = "ISO-8859-1").read()
# Lower case all text
raw_text = raw_text.lower()
# Create helper function to clean whitespace from corpus
def strip_whitespace(text,whitespace):
    _text = text
    for whtspcchar in whitespace:
        _text = _text.replace(whtspcchar,"")
    return _text
# Define list of whitespace characters
whitespace = ['\t','\n','\r','\v','\f']
# Use the helper function to strip the whitespace characters
clean_txt = strip_whitespace(raw_text,whitespace)
### This is specific to the Illiad text which ends up with some whitespace/nonsense at the beginning ###
clean_txt = clean_txt[3:]
print(clean_txt)
```

```python
clean_sentences = clean_txt.split(sep=".")
import json
sentiment = []
i=0
with open('illiad_cNLP_sentiment_test.txt', 'w') as jsonfile:
    for sentence in clean_sentences:
        output = nlp.annotate(sentence,props)
        if output == "CoreNLP request timed out. Your document may be too long.":
            next
        elif type(output) is str:
            #print(output)
            output = json.loads(output, strict=False)
            json.dump(output,jsonfile)
            jsonfile.write(',')
            
```
